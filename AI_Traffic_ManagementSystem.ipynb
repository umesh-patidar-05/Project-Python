```python
import cv2
import numpy as np
from ultralytics import YOLO
from google.colab.patches import cv2_imshow
```

```python
model = YOLO("yolov8n.pt")
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
video_path = "1.mp4"

cap = cv2.VideoCapture(video_path)
```

```python
while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    results = model(frame)

    annotated_frame = results[0].plot()

    cv2_imshow(annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
```

```python
vehicle_count = 0

for r in results:
    boxes = r.boxes

    for box in boxes:
        cls = int(box.cls[0])

        if cls in [2, 3, 5, 7]:
            vehicle_count += 1

print("Vehicles:", vehicle_count)
```

```python
if vehicle_count < 10:
    traffic = "LOW"

elif vehicle_count < 30:
    traffic = "MEDIUM"

else:
    traffic = "HIGH"

print("Traffic Density:", traffic)
```

```python
if traffic == "LOW":
    green_signal = 20

elif traffic == "MEDIUM":
    green_signal = 40

else:
    green_signal = 70

print("Green Signal Time:", green_signal)
```

```python
import cv2

cap = cv2.VideoCapture("1.mp4")

frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

fourcc = cv2.VideoWriter_fourcc(*'mp4v')

out = cv2.VideoWriter(
    '156157-811878168_tiny.mp4',
    fourcc,
    20.0,
    (frame_width, frame_height)
)

while cap.isOpened():

    ret, frame = cap.read()

    if not ret:
        break

    # yaha AI detection code chalega

    out.write(frame)

cap.release()
out.release()
cv2.destroyAllWindows()

print("Video Saved Successfully")
```
